```python
# 파일 저장하기

slice = 0
with open("./data/tube-sample.png", "rb") as f:
    byte = f.read(16)
    while byte != b"":
        # Do stuff with byte.
        byte = f.read(4096)
        if (len(byte) == 0):
            break
        decoded = np.fromstring(byte, dtype=np.uint8)
        print(decoded)
        cv2.imwrite('./data/tube-sample_{}.png'.format(i), decoded)
        slice+=1
    print(slice)
```

```python
# 파일 저장하기2 : file read stream
# reshape 필요
# # 바이트로 슬라이스한 것을 보여주기

_IMAGE_SIZE = 100
#num_images = 1
_NUM_CHANNELS = 3
sliceSize = _IMAGE_SIZE * _IMAGE_SIZE * _NUM_CHANNELS
slice = 0

with open("./data/tube-sample.png", "rb") as bytestream:
    buf = bytestream.read(16)
    while buf != b"":
        buf = bytestream.read(sliceSize)
        if len(buf) < sliceSize:
            break
        data = np.frombuffer(buf, dtype=np.uint8)
        data = data.reshape(_IMAGE_SIZE, _IMAGE_SIZE, _NUM_CHANNELS)
        print(data)
        cv2.imwrite('./data/tube-sample_{}.png'.format(slice), data)
        slice+=1
    print(slice)
```

```python
# 파일 저장하기3 : file read stream
# reshape 필요
# # 바이트로 슬라이스한 것을 보여주기

_IMAGE_SIZE = 100
width = 200
height = 100
#num_images = 1
colorChannels = 3
sliceSize = width * height * colorChannels
slice = 0

with open("./data/tube-sample.png", "rb") as bytestream:
    buf = bytestream.read(16)
    while buf != b"":
        buf = bytestream.read(sliceSize)
        if len(buf) < sliceSize:
            break
        data = np.frombuffer(buf, dtype=np.uint8)
        data = data.reshape(height, width, colorChannels)
        #print(data)
        cv2.imwrite('./data/tube-sample_{}.png'.format(slice), data)
        slice+=1
    print(slice)
```

```python
# Third party modules
import numpy
from PIL import Image


def get_image(image_path):
    """Get a numpy array of an image so that one can access values[x][y]."""
    image = Image.open(image_path, "r")
    width, height = image.size
    print(width, height)
    pixel_values = list(image.getdata())
    if image.mode == "RGB":
        channels = 3
    elif image.mode == "L":
        channels = 1
    elif image.mode == 'RGBA':
        channels = 4
    else:
        print("Unknown mode: %s" % image.mode)
        return None
    pixel_values = numpy.array(pixel_values).reshape((height,width, channels))
    return pixel_values


image = get_image("./data/tube-sample.png")
print(image.shape)
cv2.imshow('img1', image)
cv2.waitKey()
cv2.destroyAllWindows()
```
